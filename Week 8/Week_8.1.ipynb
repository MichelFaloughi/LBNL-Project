```python
# We want to find the time t at which the SOC gets to 80%
# plot sine waves ( make sure it goes from 20% to SOC_max)
```


$$
\text{SOC} = \alpha \sin(\omega t + b) + c
$$
$$
t = t(\text{SOC}_{max}, \omega)
$$

Assuming that the SOC of any cell varies between $20\%$ and a certain $\text{SOC}_{max}$, here is the proposed model:

$$ \text{SOC} = \alpha\sin(2\pi\omega + b) + c $$

$$ \text{SOC}_{max} \in [30, 100]\%, \; b = \text{constant} = 0 $$

We can clearly see that:
$$\alpha + c = \text{SOC}_{max} $$
and
$$ -\alpha + c = 20\% $$ 

Hence we have:
$$ c = \frac{\text{SOC}_{max} + 20}{2} $$
$$ \alpha = \frac{\text{SOC}_{max} - 20}{2} $$



and the proposed model simplifes to this (remember $b=0$):
$$ \text{SOC} =  (\frac{\text{SOC}_{max} - 20}{2})\sin(2\pi\omega t) + \frac{\text{SOC}_{max} + 20}{2} $$


with $\text{SOC}_{max} \in [30, 100]$  and $ \omega \in [2,8]$
$\text{SOC}_{max}$ represents the *amplitude* and $\omega$ represents the *frenquency*

and we get the time at which SOH becomes 80% as a function $t: [30, 100] \times [2, 8] \to \mathbb{R}$

Let this function be $ t(\text{SOC}_{max}, \omega) =  -0.125\text{SOC}_{max}^2 -0.2\omega^3 + 500 $
